```python
boundary = ox.geocode_to_gdf("Bengaluru, India")
boundary.to_file("data/raw/bengaluru_boundary.geojson", driver="GeoJSON")

boundary
```

```python
G = ox.graph_from_polygon(boundary.geometry.iloc[0], network_type="drive")
ox.save_graphml(G, "data/processed/roads_base.graphml")

print("Saved roads_base.graphml")
```
