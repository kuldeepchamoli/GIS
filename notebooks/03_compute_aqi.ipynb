```python
import requests
import json
import os

# Ensure directory exists
os.makedirs("data/raw", exist_ok=True)

# Bengaluru Coordinates
lat = 12.9716
lon = 77.5946

url = (
    "https://air-quality-api.open-meteo.com/v1/air-quality?"
    f"latitude={lat}&longitude={lon}"
    "&hourly=pm10,pm2_5,carbon_monoxide,nitrogen_dioxide,sulphur_dioxide,ozone"
)

print("Fetching AQI from Open-Meteo...")

response = requests.get(url)
response.raise_for_status()

aqi_json = response.json()

```

```python
import requests
import json

url = "https://api.openaq.org/v2/latest?city=Bengaluru"

response = requests.get(url)

# safety check
response.raise_for_status()

aqi_json = response.json()

# save to file
with open("data/raw/openaq_raw.json", "w") as f:
    json.dump(aqi_json, f, indent=2)

print("Saved: data/raw/openaq_raw.json")
```
